```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.xlim(-5, 20)
plt.ylim(-2, 2);
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), color = 'green', linestyle='dashed')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), color = 'g', linestyle='--')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'g--')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'g--')
plt.plot(x, np.cos(x), 'b-+')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), 'g--')
plt.plot(x, np.cos(x), 'b-+')
plt.xlim(-1, 12)
plt.ylim(-2, 2);
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.show()
```
